## Introduction

<p>
Alors que le monde entrait dans l'ère du Big Data, le besoin de stockage augmentait également. C'était le principal défi et la principale préoccupation pour les industries de l'entreprise jusqu'en 2010. L'objectif principal était de construire des frameworks et des solutions pour stocker les données.
</p>
<p>Maintenant, que Hadoop et d'autres frameworks ont résolu avec succès le problème de stockage, l'accent a été mis sur le traitement de ces données. Data Science est la sauce secrète ici. Toutes les idées que vous voyez dans les films de science-fiction d'Hollywood peuvent réellement devenir réalité par Data Science. Data Science est le futur de l'intelligence artificielle. Par conséquent, il est très important de comprendre ce qu'est <strong>Data Science</strong> et comment cela peut ajouter de la valeur à votre entreprise.
</p>
<p>
Dans ce chapitre nous couvrirons les sujets suivants :
</p>
<ul>
<li>Le besoin de Data Science.
<li>Qu'est-ce que Data Science?
<li>En quoi est-ce différent de Business Intelligence (BI) et d'analyse de données?
<li>Le cycle de vie de Data Science à l'aide d'un cas d'utilisation.
</ul>
<p>
À la fin de ce chapitre, vous serez en mesure de comprendre ce qu'est Data Science et son rôle dans l'extraction d'informations significatives à partir des ensembles complexes et volumineux de données qui nous entourent.
</p>

## Pourquoi le data Science ?

<p>
Traditionnellement, les données que nous avions étaient principalement structurées et de petite taille, ce qui pouvait être analysé en utilisant les simples outils de Business Intelligence (BI). Contrairement aux données des systèmes traditionnels qui étaient principalement structurées, la plupart des données sont aujourd'hui non structurées ou semi-structurées. 
</p>
<p>
Voyons  les tendances des données dans la figure ci-dessous. la figure montre que d'ici 2020, plus de 80% des données seront non structurées.
</p>
<img src="images/fig1.png">
<p>
Ces données sont générées à partir de différentes sources telles que les journaux financiers, les fichiers texte, les formulaires multimédias, les capteurs et les instruments. Les outils de BI simples ne sont pas capables de traiter ce volume énorme et la variété des données. C'est pourquoi nous avons besoin d'outils analytiques et d'algorithmes plus complexes et plus avancés pour <i>traiter</i>, <i>analyser</i> et <i>tirer des conclusions significatives</i>.
</p>
<p>
Ce n'est pas la seule raison pour laquelle Data Science est devenu si populaire. Voyons comment le Data Science est utilisé dans divers domaines :
</p>
<ul>
<li>Vous pouvez comprendre les exigences précises de vos clients à partir des données existantes telles que l'historique de navigation, l'historique des achats, l'âge et le revenu du client. Ces données existaient dans le passé, mais maintenant, avec la quantité et la variété des données, vous pouvez former des modèles plus efficacement et recommander le produit à vos clients avec plus de précision. Ne serait-ce pas incroyable, car cela apportera plus de business à votre organisation?
<li>Prenons un scénario différent pour comprendre le rôle du Data Science dans la prise de décision. Et si votre voiture avait l'intelligence de vous ramener à la maison ? Les voitures autonomes recueillent des données en direct à partir de capteurs, y compris des radars et des caméras pour créer une carte de ses environs. Sur la base de ces données, elles prennent des décisions comme quand accélérer, quand décélérer, quand dépasser, où prendre un virage - en utilisant des algorithmes avancés de machine learning (apprentissage automatique).
<li>Voyons comment le Data Science peut être utilisé dans l'analyse prédictive. Prenons la prévision météorologique comme exemple. Les données des navires, des avions, des radars, des satellites peuvent être collectées et analysées pour construire des modèles. Ces modèles permettront non seulement de prévoir le temps, mais aussi de prévoir l'occurrence de catastrophes naturelles. Cela vous aidera à prendre des mesures appropriées à l'avance et à sauver de nombreuses vies.
</ul>

## Qu'est-ce que le Data Science?

<p>
L'utilisation du terme Data Science est de plus en plus courante, mais que signifie ce terme exactement ? De quelles compétences avez-vous besoin pour devenir un Data Scientist ? Quelle est la différence entre BI et Data Science ? Comment sont prises les décisions et les prédictions dans Data Science? Ce sont quelques-unes des questions qui seront traitées dans la suite de ce chapitre.
</p>
<p>
Commençons par ce qu'est le terme de Data Science. Data Science est un mélange de divers outils, algorithmes et principes de machine learning ayant pour but de déduire des modèles à partir des données brutes (raw data en anlais). Mais, en quoi est-ce différent de ce que les statisticiens font depuis des années ?
</p>
<p>
La réponse réside dans la différence entre <strong>expliquer</strong> et <strong>prédire</strong>.
</p>
<p>
<img src="images/fig3.png">
</p>
<p>
Comme vous pouvez le voir à partir de la figure ci-dessus, un analyste de données (data analyst) explique généralement ce qui se passe en traitant l'historique des données. D'un autre côté, un Data Scientist fait non seulement l'analyse exploratoire pour en découvrir les intuitions, mais utilise également divers algorithmes avancés de machine learning pour identifier l'occurrence d'un événement particulier dans le futur. Un Data Scientist examinera les données sous plusieurs angles, parfois des angles inconnus auparavant.
</p>
<p>
Ainsi, le Data Science est principalement utilisé pour prendre des décisions et des prédictions en utilisant l'analyse prédictive, l'analytique prescriptive et le machine learning.
</p>

<ul>
<li><strong>Analyse prédictive :</strong> Si vous voulez un modèle capable de prédire les possibilités d'un événement particulier à l'avenir, vous devez appliquer une analyse prédictive. Par exemple, quelle est la probabilité pour que les clients effectuent des remboursement de leurs crédits. Ici, vous pouvez construire un modèle qui peut effectuer des analyses prédictives sur l'historique de paiement des clients pour prédire si les futurs paiements seront à temps ou non.
<li><strong>Analyse prescriptive :</strong> Si vous voulez un modèle qui a l'intelligence de prendre ses propres décisions et la capacité de le modifier avec des paramètres dynamiques, vous avez certainement besoin d'une analyse prescriptive. Ce domaine relativement nouveau consiste à fournir des conseils. En d'autres termes, non seulement il prédit mais suggère une gamme d'actions prescrites et des résultats associés. Le meilleur exemple pour cela est la voiture auto-conduite de Google. Les données recueillies par les véhicules peuvent être utilisées pour former des voitures autonomes. Vous pouvez exécuter des algorithmes sur ces données pour y apporter de l'intelligence. Cela permettra à votre voiture de prendre des décisions comme quand tourner, quel chemin prendre, quand ralentir ou accélérer.
<li><strong>Le machine learning pour faire des prédictions :</strong> Si vous avez des données transactionnelles d'une société de financement et que vous avez besoin de construire un modèle pour déterminer la tendance future, alors les algorithmes de machine learning sont le meilleur choix. Cela relève du paradigme de l'apprentissage supervisé. Il est appelé supervisé parce que vous avez déjà les données sur lesquelles vous pouvez former vos machines. Par exemple, un modèle de détection de fraude peut être formé en utilisant un historique des achats frauduleux.
<li><strong>Le machine learning pour la découverte de modèles :</strong> Si vous n'avez pas les paramètres sur lesquels vous pouvez vous basez pour faire des prédictions, vous devez rechercher des modèles dans l'ensemble de données pour pouvoir faire des prédictions significatives. Ce n'est rien d'autre que le modèle non supervisé car vous n'avez aucune étiquette prédéfinie pour le regroupement. L'algorithme le plus couramment utilisé pour la découverte de modèles est le "Clustering".

</ul>

## Business Intelligence (BI) et Data Science

<p>
Souvent, le terme de Data Science est confondu avec celui Business Intelligence. Voici quelques points pour comprendre les différences entres ces deux concepts :
</p>
<ul>
<li>BI analyse essentiellement les données précédentes pour trouver le recul et la perspicacité pour décrire les tendances de l'entreprise. BI vous permet de prendre des données provenant de sources externes et internes, de les préparer, d'exécuter des requêtes et de créer des tableaux de bord pour répondre aux questions telles que l'analyse des revenus trimestriels ou les problèmes métiers. BI peut évaluer l'impact de certains événements dans un proche avenir.
<li>Data Science est une approche plus prospective, une approche exploratoire axée sur l'analyse des données du passé ou actuelles et la prédiction des résultats futurs dans le but de prendre des décisions éclairées. Il répond aux questions ouvertes sur les événements «quoi» et «comment».
</ul>

## Cycle de vie de Data Science

<p>
Une erreur courante commise dans les projets Data Science est de se précipiter dans la collecte et l'analyse de données, sans comprendre les exigences ou même d'encadrer correctement le problème métier. Par conséquent, il est très important de suivre toutes les phases du cycle de vie de Data Science pour assurer le bon fonctionnement du projet.
</p>
<p>
Voici un bref aperçu des principales phases du cycle de vie de Data Science :
</p>
<p>
<img src="images/fig4.png" width="50%">
</p>

<ul>
<li><strong>Phase 1 – Découverte (Discovery) :</strong> Avant de commencer le projet, il est important de comprendre les diverses spécifications, exigences, priorités et budget requis. Vous devez posséder la capacité de poser les bonnes questions. Dans cette pahse, vous évaluez si vous disposez des ressources nécessaires en termes de personnes, de technologie, de temps et de données pour soutenir le projet. Vous devez également encadrer le problème métier et formuler les hypothèses initiales (IH) à tester.
<li><strong>Phase 2 - Préparation des données (Data preparation):</strong> Dans cette étape, vous devez explorer, pré-traiter et conditionner les données avant la modélisation. De plus, vous allez effectuer ETLT (extraire, transformer, charger et transformer) pour obtenir des données prêt à utiliser.
</ul>
<p>
<img src="images/fig5.png">
</p>

<p>
Vous pouvez utiliser divers outils, tels que R ou Python pour le nettoyage, la transformation et la visualisation des données. Cela vous aidera à repérer les valeurs aberrantes et à établir une relation entre les variables. Une fois que vous avez nettoyé et préparé les données, il est temps de faire des analyses exploratoires. Voyons voir comment vous pouvez y parvenir.
</p>

<ul>
<li><strong>Phase 3 - Planification du modèle (Model planning):</strong> Ici, vous allez déterminer les méthodes et les techniques pour déduire les relations entre les variables. Ces relations définiront la base des algorithmes que vous implémenterez dans la phase suivante. Vous appliquerez l'analyse des données exploratoires (EDA) en utilisant diverses formules statistiques et outils de visualisation.
</ul>
<p>
La figure suivante montre divers outils de planification de modèle :
</p>

<p>
<img src="images/fig6.jpg">
</p>
<ol>
<li>R possède un ensemble complet de capacités de modélisation et fournit un bon environnement pour construire des modèles d'interprétation.
<li>Les services d'analyse SQL peuvent effectuer des analyses dans la base de données à l'aide de fonctions d'exploration de données communes et de modèles prédictifs de base.
<li>SAS / ACCESS peut être utilisé pour accéder aux données de Hadoop et être utilisé pour créer des diagrammes de flux de modèle réutilisables.
</ol>

<p>
Bien que de nombreux outils soient présents sur le marché, R est l'outil le plus utilisé.
</p>

<p>
Maintenant que vous avez un aperçu sur la nature de vos données et avez décidé des algorithmes à utiliser, vous allez appliquer, dans la prochaine étape, un algorithme et construire un modèle.
</p>
<ul>
<li><strong>Phase 4 - Construction du modèle (Model building):</strong> Au cours de cette phase, vous développerez des ensembles de données à des fins de formation et de test. Vous examinerez si vos outils existants suffiront pour exécuter les modèles ou si vous avez besoin d'un environnement plus robuste (comme le traitement parallèle). Vous analyserez différentes techniques d'apprentissage telles que la classification, l'association et le regroupement pour construire le modèle.
</ul>

<p>
Vous pouvez réaliser le modèle en utilisant les outils suivants :
</p>

<p>
<img src="images/fig7.jpg">
</p>

<ul>
<li><strong>Phase 5 - Opérationnaliser (operationalize):</strong> Dans cette phase, vous livrez des rapports finaux, des briefings, du code source et des documents techniques. En outre, un projet pilote est parfois mis en œuvre dans un environnement de production en temps réel. Cela vous fournira une image claire de la performance et d'autres contraintes connexes à petite échelle avant le déploiement complet.
</ul>

<ul>
<li><strong>Phase 6 - Communiquer les résultats (communicate results):</strong> Maintenant, il est important d'évaluer si vous avez été en mesure d'atteindre votre objectif que vous aviez prévu dans la première phase. Ainsi, dans la dernière phase, vous identifiez tous les résultats clés, communiquez aux parties prenantes et déterminez si les résultats du projet sont un succès ou un échec basé sur les critères développés dans la phase 1.
</ul>

<strong>Étape 1 :</strong>

<ul>
<li>Comme indiqué dans la phase 1 du cycle de vie, nous allons commencer tout d'abord par collecter les données en fonction des antécédents médicaux du patient. Vous pouvez vous référer aux exemples de données ci-dessous.

Ces données présentent de nombreuses incohérences, telles que des valeurs manquantes, des colonnes vides, des valeurs abruptes et un format de données incorrect. Toutes ces incohérences doivent être nettoyées.

<li>Dans la colonne <strong>bp</strong>, l'une des valeurs est <strong>6600</strong>, ce qui est impossible (du moins pour les humains) car la pression artérielle ne peut pas atteindre une telle valeur.

<p>

<li>Comme nous avons les principaux attributs pour l'analyse comme <strong>npreg, bmi, etc.</strong>, nous allons utiliser la technique d'apprentissage supervisé pour construire un modèle.

Dans notre cas, nous avons une relation linéaire entre <strong>npreg</strong> et <strong>age</strong>, alors que la relation est non linéaire entre <strong>npreg</strong> et <strong>ped</strong>.
<li>Les modèles d'arbre de décision sont également très robustes car nous pouvons utiliser la combinaison de différentes d'attributs pour faire des arbres différents et finalement mettre en œuvre celui avec l'efficacité maximale.
