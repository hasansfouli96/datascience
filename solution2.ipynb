```python
import pandas as pd
import matplotlib


#Partie 1
#Question1
def retourneAnnee(chaine):
    result=chaine[-4:]
    return(result)

print(retourneAnnee("Tue Oct 18 21:53:25 +0000 2011"))

#Question2

def nbTweetsParAnnee(fichier):
    d=dict()
    f=open(fichier,encoding="latin-1")  
    data=f.readlines()
    for i in range(15):
        if(i!=0):
            ligne=data[i].split('","')
            date=retourneAnnee(ligne[3])
            if date in d:
                nb=d[date]
                d[date]=nb+1
            else:
                d[date]=1       
    return d
    

print(nbTweetsParAnnee("tweets-corpus.csv"))


#Question3
def pourcentageDeTweets(fichier):
    d=dict()
    f=open(fichier,encoding="latin-1")  
    data=f.readlines()
    for i in range(29):
        if(i!=0):
            ligne=data[i].split('","')
            if ligne[1]=="positive":
                if "positive" in d:
                    nb=d["positive"]
                    d["positive"]=nb+1
                else:
                    d["positive"]=1  
            
            elif ligne[1]=="neutral":
                if "neutral" in d:
                    nb=d["neutral"]
                    d["neutral"]=nb+1
                else:
                    d["neutral"]=1  
                    
            elif ligne[1]=="irrelevant":
                if "irrelevant" in d:
                    nb=d["irrelevant"]
                    d["irrelevant"]=nb+1
                else:
                    d["irrelevant"]=1 
    for i in d:
        nb=d[i]
        d[i]=nb/5112
    return d
print(pourcentageDeTweets("tweets-corpus.csv"))
```
